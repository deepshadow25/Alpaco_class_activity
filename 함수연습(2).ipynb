```python
import re
```

```python
def hangul(text):
    return re.sub('[^가-힣]', '', text)  
# sub 함수 복습 필요
# hangul(text) : text값을 넣어서 hangul 함수를 실행(return)시키겠다.
```

```python
hangl_text = []

hangl_text.append(hangul('abc가나다345'))
hangl_text
#result는 hangl 함수에 text값을 넣은 결과로 정의하겠다.
```

```python
def remtwo(n):
    result = n%2
    return result
```

```python
result = remtwo(7)
result # remtwo 함수에 n(=7)을 넣었을 때의 결과를 return하여 result라는 새로운 변수로 정의하겠다.
```
