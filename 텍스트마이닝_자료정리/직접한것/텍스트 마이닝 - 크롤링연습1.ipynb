검색어에 따른 네이버 뉴스기사 크롤링
- 검색어 : 시진핑

```python
news_xi = {
    "제목" : [], 
    "주소" : []
}

for i in range(10):

    url4 = 'https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EC%8B%9C%EC%A7%84%ED%95%91&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=50&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start={}'.format(i)
    resp4 = requests.get(url4)
    
    html4 = resp4.content
    soup4 = BeautifulSoup(html4, 'lxml')
    a_tags4 = soup4.select('a.news_tit')

    for a_tag in a_tags4:
        news_xi["제목"].append(a_tag['title'])
        news_xi["주소"].append(a_tag['href'])
```

```python
import pandas as pd
df_xi = pd.DataFrame(news_xi)
df_xi
df_xi.to_excel('df_xi.xlsx', index=False)
```

네이버 웹툰 제목, 작가, 평점, 요일 크롤링
- 현재 연재중인 웹툰만.

```python
url = 'https://comic.naver.com/webtoon/weekdayList?week=mon'
week_links = []

resp = requests.get(url)
html = resp.content
soup = BeautifulSoup(html, 'lxml')
a_tags = soup.select('ul.category_tab a')

for idx, a_tag in enumerate(a_tags):
    if idx:
        week_links.append('https://comic.naver.com' + a_tag['href'])  

week_links
```

```python
a_tags = soup.select('ul.category_tab a')[1:]

weeks = []
for a_tag in a_tags:
    weeks.append(a_tag.text[:-3])
weeks
```

```python
webtoon_lists = {
    "제목" : [],
    "작가" : [],
    "평점" : [],
    '요일' : []
}

weeks = []

for url in week_links:
    week = url.split('=')[-1]

    a_tags = soup.select('ul.category_tab a')[1:]


    resp = requests.get(url)
    html = resp.content
    soup = BeautifulSoup(html, 'lxml')

    tit_tags = soup.select('ul.img_list li dl dt a')
    aut_tags = soup.select('ul.img_list dd.desc a') # 클래스가 2개 이상 - 클래스 사이만 징검다리로 건너도 안정적
    rat_tags = soup.select('div.rating_type strong')
    week_tags = soup.select('ul.category_tab')[1:]
    
    for tit_tag, aut_tag, rat_tag in zip(tit_tags, aut_tags, rat_tags):
        webtoon_lists['제목'].append(tit_tag['title'])
        webtoon_lists['작가'].append(aut_tag.text)
        webtoon_lists['평점'].append(rat_tag.text)
        webtoon_lists['요일'].append(week)

webtoon_lists = pd.DataFrame(webtoon_lists)
webtoon_lists
```
