```python
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
from tqdm import tqdm
import json
```

```python
url = 'https://www.google.com/search?q=%EA%B0%95%EC%95%84%EC%A7%80&source=lnms&tbm=isch&sa=X&ved=2ahUKEwjngP6jqon7AhX27zgGHcvUCucQ_AUoAXoECAIQAw&biw=639&bih=600&dpr=1.5&sfr=gws&gbv=1&sei=WChfY9vaNMXU0wT335aICQ'

resp = requests.get(url)

soup = BeautifulSoup(resp.content, 'lxml')
img_tags = soup.select('img.yWs4tf')
len(img_tags)

for idx, img_tag in enumerate(img_tags):
    img_src = img_tag['src']
    resp_img = requests.get(img_src)

    img = resp_img.content

    with open('강아지_{:02}.jpg'.format(idx+1), 'wb') as f:
        f.write(img)
```

```python
for i in tqdm(range(10), desc='페이지'):
    start = 20*i
    url = 'https://www.google.com/search?q=%EA%B0%95%EC%95%84%EC%A7%80&biw=639&bih=600&gbv=1&tbm=isch&ei=WShfY7TFJYiB2roP6a2_kA0&start={}&sa=N'.format(start)
    resp = requests.get(url)

    soup = BeautifulSoup(resp.content, 'lxml')
    img_tags = soup.select('img.yWs4tf')
    len(img_tags)

    for idx, img_tag in tqdm(enumerate(img_tags), desc='이미지/페이지'):
        img_src = img_tag['src']
        resp_img = requests.get(img_src)

        img = resp_img.content

        with open('강아지_{:03}.jpg'.format(start+idx+1), 'wb') as f:
            f.write(img)
```

### 요청이 거절당했을 때

1. 보내는 사람(Requests header - Client info) //
받는 사람(Responce header - Server info)

2. 브라우저는 OK, 크롤러(견)는 403. 브라우저의 Requests header는 OK, 크롤러는 거절.

3. 브라우저의 Requests header를 복붙해서 GET하자.

4. requests header 정보 중에서
- 1순위 user-agent, referrer
- 2순위 cookies
- 3순위 auth, authorization 등 권한을 의미하는 단어가 포함된 값.
- 4순위 전부 다

```python
url = 'https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json'

params = {
    'ticket' : 'comic',
    'templateld' : 'webtoon',
    'pool' : 'cbox3',
    '_cv' : 20221018144728,
    '_callback' : '', #jQuery112405450122901261651_1667195002901
    'lang' : 'ko',
    'country' : 'KR',
    'objectId' : '783054_13',
    'pageSize' : 15,
    'indexSize' : 10,
    'listType' : 'OBJECT',
    'pageType' : 'default',
    'page' : 1,
    'refresh' : 'true',
    'sort' : 'new'
}

# ?ticket=comic&templateId=webtoon&pool=cbox3&_cv=20221018144728
# &_callback=jQuery112405450122901261651_1667195002901&lang=ko&country=KR&objectId=783054_13
# &categoryId=&pageSize=15&indexSize=10&groupId=&listType=OBJECT&pageType=default
# &page=1&currentPage=1&refresh=true&sort=new&followSize=&_=1667195002906


headers = {'referer': 'https://comic.naver.com/comment/comment?titleId=783054&no=13', 
          'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
resp = requests.get(url, params=params, headers=headers)
resp
```

```python
resp.text
```

```python
result = resp.text.replace('_callback(', '')
result = result.replace(');', '')
result = json.loads(result)
result
```

```python
contents, sympath, antipath = [], [], []

last_page = result['result']['pageModel']['totalPages']
last_page
```

```python
commentlists = result['result']['commentList']

for page in tqdm(range(last_page)):

    for commentlist in commentlists:
        contents.append(commentlist['contents'])
        sympath.append(commentlist['sympathyCount'])
        antipath.append(commentlist['antipathyCount'])
        
commentListTotal = {
    "댓글" : contents,
    "공감" : sympath,
    "비공감" : antipath
}
```

```python
reviews = {
    '댓글' : [],
    '공감수' : [],
    '비공감수' : [],
    '화' : []
}
last_no = 20
for part in tqdm(range(1, last_no+1)):
    url = 'https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json'

    params = {
        'ticket' : 'comic',
        'templateld' : 'webtoon',
        'pool' : 'cbox3',
        '_callback' : '#jQuery112405450122901261651_1667195002901',
        'lang' : 'ko',
        'country' : 'KR',
        'objectId' : '783054_{}'.format(part),
        'pageSize' : 15,
        'indexSize' : 10,
        'listType' : 'OBJECT',
        'pageType' : 'default',
        'page' : 1,
        'refresh' : 'true',
        'sort' : 'new'
    }

    headers = {'referer': 'https://comic.naver.com/comment/comment?titleId=783054&no=13', 
          'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}

    resp = requests.get(url, params=params, headers=headers)

    jsn = resp.text[43:-2]

    dict_resp = json.loads(jsn)

    commentList = dict_resp['result']['commentList']

    for comment in commentList:
        reviews['댓글'].append(comment['contents'])
        reviews['공감수'].append(comment['sympathyCount'])
        reviews['비공감수'].append(comment['antipathyCount'])
        reviews['화'].append(part)
```

```python
reviews = pd.DataFrame(reviews)
reviews.to_csv('쇼미더럭키짱_전체댓글.csv')
```
