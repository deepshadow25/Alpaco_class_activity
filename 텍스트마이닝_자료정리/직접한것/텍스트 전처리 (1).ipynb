```python
import re
```

```python
text = 'qweRTY123456가나다라마바1'
re.findall('3.+?',text)
```

```python
text = 'alpaco123@gmail.com'
re.findall('[0-9].+?', text) # [0~9]+  숫자1회이상 : 숫자열
```

```python
text = '신라면이 집에 있다면 먹고 싶다.'
re.findall('신.+?면', text)
# .+ (greedy) : 신라면이 집에 있다'면'
# .+? (reluctant) : 신라'면'
```

### 문서에서 특정 대상을 찾아내고플때 : 정규표현식 마렵다

정규표현식을 이용해 특정 대상을 어떻게 표현할 것인가
- 1. 대상의 내부적 특징을 표현할 것인지
- 2. 대상의 외부적(위치적) 특징을 표현할 것인지(,+, ,+?)

사실, 내/외부 모두 고려하게 됨.

우리끼리만 일치하는 공통적 특징 찾기!
- 내부적 특징이든 외부적 특징이든 자신에게 편한 쪽부터 먼저 찾는다.
- 가능여부 판단 후, 가능하다면 한쪽만 표현하고 끝, 불가능하다면 다른 쪽까지

####다소 구차할지라도 실수하지 않는 것이 가장 중요!

```python
text = '''1. 삼성전자 100,000
2. 현대차 300,000
3. 네이버 1,000,000'''
re.findall('([가-힣].+?\s)', text)  # [가-힣]\s : 한글 뒤에 있는 숫자열([0-9].+)을 찾아냄.
```

토큰의 종류는 학습해야 할 자료의 수다. (데이터 개수는 아님)

토큰의 종류가 많다 : 학습할 내용이 많다는 것.

학습 난이도가 너무 낮아도 실전 성능이 안 나올 것.

'적절히' 학습 난이도를 낮추는 것이 중요.

: 필요한 자료를 최대한 뽑아 쓰겠다
