```python
class Account:
    num_account = 0
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.bankname = 'SC은행'
        
        a = random.randint(100, 999)
        b = random.randint(10, 99)
        c = random.randint(100000, 999999)
        self.AcNum = f'{a}-{b}-{c}'

        Account.num_account +=1
        self.deposit_time = 0
        self.withdraw_time = 0

    def get_account_num(self):
        return self.num_account
    def deposit(self, money):
        self.deposit_time += 1
        print(f'{money}원 입금되었습니다.')
        if money > 0:
            self.balance += money
            if self.deposit_time%5 == 0 and self.deposit_time >0:
                self.balance += self.balance * 1.01
        else:
            raise ValueError
        

    def withdraw(self, money):
        self.withdraw_time += 1
        print(f'{money}원 출금되었습니다.')
        self.balance -= money
        if self.balance < 0:
            print('잔액부족')
            raise ValueError


    def display_info(self):
        print(f'은행이름:{self.bankname}')
        print(f'예금주:{self.name}')
        print(f'계좌번호:{self.AcNum}')
        print(f'잔액:{self.balance}')

    def deposit_history(self):
        return self.deposit_time

    def withdraw_history(self):
        return self.withdraw_time
```

```python
with open('매수종목.csv', 'wt', encoding='cp949') as scsv:
    scsv.write('''종목명 [삼성전자 Naver] 
                종목코드 [005930 035240]
                PER [15.79, 55.82]''')

with open('매수종목.csv','rt', encoding='cp949') as scsv:
    sclist = scsv.read().split()
    scdickey = []
    scdicvalue = []
    for i in range(len(sclist)):
        if i%2 == 0:
            scdickey.append(sclist[i])
        else:
            scdicvalue.append(sclist[i])
scdic = dict(zip(scdickey, scdicvalue))
scdic
```
