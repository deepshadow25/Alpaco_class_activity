## 문제 11) 타이타닉 데이터프레임에 "성별2" 열을 추가하세요.
#### * "성별2" 열은 "남자", "여자"로 구성되지만, 나이가 20살 미만이라면 "어린이" 라는 값으로 값을 구성해주세요.



```python
sex = titanic['sex']

for i in range(0,890):
    if titanic['age'][i] < 20 :
        sex2 = sex.replace(sex[i],'어린이')
titanic['성별2'] = sex2
titanic
```
