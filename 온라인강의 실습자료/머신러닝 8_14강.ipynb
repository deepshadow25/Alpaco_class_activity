```python
# 10강  특성공학을 사용하여 특성 늘리기

import pandas as pd

perch = pd.read_csv('perch.csv')
target = perch['weight'].to_numpy()
perch2 = pd.read_csv('perch2.csv').to_numpy()
perch2
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(include_bias=False)
perch_poly = poly.fit_transform(perch2)
perch_poly.shape  # x1, x2, x3, 각각의 제곱, 서로다른 두 특성을 곱한 값 생성
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(perch_poly, target)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
train_scaled = ss.fit_transform(train_input) #train_input의 평균, 표준편차를 학습하여 얻고 변환하여 넣어라
test_scaled = ss.transform(test_input) #위에서 구한 train_input의 평균, 표준편차로 변환하여 test_scaled에 넣어라
```

```python
test_scaled
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(train_input, train_target)
```

```python
lr.score(test_input, test_target)
```

```python
# 12강  트리 기반 모델

from sklearn.datasets import load_iris

iris = load_iris()
data, target = iris['data'], iris['target']
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(data, target)
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree

dt = DecisionTreeClassifier(max_depth=3)  #과대적합방지
dt.fit(train_input, train_target)
dt.score(test_input, test_target) 
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plot_tree(dt, feature_names=iris['feature_names'], filled=True)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(train_input, train_target)
rf.score(test_input, test_target)
```

```python
# 13강  교차검증 
# 검산과정과 유사함(시험데이터로만 성능검증을 하면 시험데이터에만 잘 맞는 모델이 되어버리기에 필요한 과정)
# 검증 데이터를 따로 만들어서 조정하고 최종적으로 시험데이터에서 최종검증.

from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, train_test_split

iris = load_iris()
data, target = iris['data'], iris['target']

train_input, test_input, train_target, test_target = train_test_split(data, target)

rf = RandomForestClassifier()
scores = cross_val_score(rf, data, target, cv=5)
```

```python
import numpy as np

np.mean(scores)
```

```python
rf.fit(train_input, train_target)
```

```python
rf.score(test_input, test_target)
```

```python
# 14강  시뮬레이션을 통한 최적의 머신러닝 모델 만들기

# Grid Search (격자탐색) : 가장 고성능, 그러나 시간이 매우 많이걸림

import pandas as pd
from sklearn.datasets import load_wine

wine = load_wine()
```

```python
df = pd.DataFrame(wine['data'], columns=wine['feature_names'])
df['targer'] = wine['target']
df
```

```python
wine['target_names']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

data, target = wine['data'], wine['target']
```

```python
rf = RandomForestClassifier()

dic = {'n_estimators' : range(100, 201, 10), "max_depth" : range(1,11)}

gs = GridSearchCV(estimator=rf, param_grid=dic,cv=3)
gs.fit(data, target)  
```

```python
gs.best_params_
```

```python
result = gs.cv_results_
df = pd.DataFrame(result)
df.to_excel('result.xlsx')
```
