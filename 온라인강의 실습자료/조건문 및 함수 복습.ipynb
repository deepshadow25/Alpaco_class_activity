```python
#for문을 이용한 팩토리얼함수 만들기

n=int(input("팩토리얼을 구할 수를 입력하세요 : "))
result = 1
for item in range(1, n+1, 1):
    result *= item   # result = result * item
if n<0:
    raise ValueError
print(result)

import math   # math 라이브러리를 이용한 팩토리얼 구하기
math.factorial(n)
```
