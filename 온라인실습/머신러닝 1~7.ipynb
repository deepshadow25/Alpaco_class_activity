```python
length = fish['length'].to_numpy()
weight = fish['weight'].to_numpy()
```

```python
data5 = []

for i in range(49):
    data5.append([length[i], weight[i]])

data5 = np.array(data5)
target5 = fish['class'].to_numpy()
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()   # 모델생성
knn.fit(data5, target5)
```

```python
knn.predict([[35, 400]])    # 기본 : 5개 가져옴
```

```python
for i in range(1, 50, 2):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(data5, target5)
    score = knn.score(data5, target5)
    print("이웃수 : {} // 점수 : {}".format(i, score))
```

```python
data = fish[['length', 'weight']].to_numpy()
target = fish['class'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

train_input, test_input, train_target, test_target = train_test_split(data, target)

knn = KNeighborsClassifier()
knn.fit(train_input, train_target)
knn.score(test_input, test_target)
```

```python
knn.predict([[25, 150]])
```

```python
import numpy as np

mean = np.mean(train_input, axis=0) # 행방향
std = np.std(train_input, axis=0)

train_scaled = (train_input - mean) / std  # 훈련데이터 (학습한 데이터 : 기준치)
test_scaled = (test_input - mean) / std   # 시험데이터 (모르는 데이터 : 모르는 데이터를 잘 예측, 분석해야 좋은 프로그램)
```

```python
knn = KNeighborsClassifier()
knn.fit(train_scaled, train_target)
knn.score(test_scaled, test_target)
```

```python
new = ([25, 150] - mean) /std
new
```

```python
knn.predict([new])
```

```python
# 스케일링 라이브러리 추가학습 필요

from sklearn.preprocessing import StandardScaler

standardscaler = StandardScaler().fit(train_scaled)
X_train_standard=standardscaler.transform(train_scaled)
X_test_standard=standardscaler.transform(test_scaled)
print(standardscaler, X_train_standard, X_test_standard)
```
