```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

anscombe = sns.load_dataset("anscombe")
anscombe
```

```python
data1 = anscombe[anscombe['dataset'] == 'I']
data2 = anscombe[anscombe['dataset'] == 'II']
data3 = anscombe[anscombe['dataset'] == 'III']
data4 = anscombe[anscombe['dataset'] == 'IV'] 

data4
```

```python
fig=plt.figure() # 그림을 그리기 위한 도화지

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
```

```python
ax1.scatter(data1['x'], data1['y'])
ax2.scatter(data2['x'], data2['y'])
ax3.scatter(data3['x'], data3['y'])
ax4.scatter(data4['x'], data4['y'])

fig
```

```python
ax1.plot(data1['x'], data1['y'], 'green')
ax2.plot(data2['x'], data2['y'], 'green')
ax3.plot(data3['x'], data3['y'], 'red')
ax4.plot(data4['x'], data4['y'], 'red')

fig
```

```python
ax1.set_title("Data1")
ax2.set_title("Data2")
ax3.set_title("Data3")
ax4.set_title("Data4")

fig.tight_layout()
fig
```

```python
# 25강 matplot

tips=sns.load_dataset("tips")
tips
```

```python
fig=plt.figure()

ax1 = fig.add_subplot(1,1,1)
ax1.hist(tips['total_bill'], bins=5)
```

```python
fig=plt.figure()

ax1 = fig.add_subplot(1,1,1)
ax1.scatter(tips['total_bill'], tips['tip'], c=tips['size']*20, alpha=0.5)
ax1.set_xlabel("Total Bill")
ax1.set_ylabel("Tip")
ax1.set_title("Scatter")
```

```python
fig=plt.figure()

ax1 = fig.add_subplot(1,1,1)
# 1,1,1 : 하나만 그리므로 하나의 그래프만 있어야 함.

male = tips[tips['sex'] == 'Male']
female = tips[tips['sex'] == 'Female']
female

ax1.boxplot([male['total_bill'], female['total_bill']])
```

```python
# 26강 seaborn

tips = sns.load_dataset("tips")
tips
```

```python
sns.histplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'], tips['tip'], shade=True)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex ')
```

```python
sns.boxplot(x='sex', y='total_bill', data=tips)
```

```python
sns.violinplot(x='time', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.lmplot(x='x', y='y', data=anscombe, col = 'dataset', col_wrap = 2)
```

```python
sns.set_style("whitegrid")
facet = sns.FacetGrid(tips, col='time', hue='sex', row='smoker')
facet.map(plt.scatter, 'total_bill', 'tip')
```
